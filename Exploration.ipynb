La apasionante busqueda de como obtenemos la imagen rgb

```python
lr_image_dir = "E:\TFG\core_dataset\lr_dataset\Amnesty POI-1-1-1\L2A\\"
filenames = os.listdir(lr_image_dir)
images = []
for filename in filenames:
    if filename.endswith(".tiff") and filename.__contains__("1-1-1-1"):
        print(filename)
        images.append(tifffile.imread(lr_image_dir + filename))
```

```python
for image in images:
    print(image.shape)
```

Todos tienen la misma forma pero solo el "data" tiene m√°s de 1 canal. Ahora lo intentaremos visualizar

```python
blue_band = images[3][:,:,0]
green_band = images[3][:,:,1]
red_band = images[3][:,:,2]

rgb_image = np.zeros((images[3].shape[0], images[3].shape[1], 3))
rgb_image[:,:,0] = red_band
rgb_image[:,:,1] = green_band
rgb_image[:,:,2] = blue_band
```

```python
tifffile.imshow(rgb_image)
```
